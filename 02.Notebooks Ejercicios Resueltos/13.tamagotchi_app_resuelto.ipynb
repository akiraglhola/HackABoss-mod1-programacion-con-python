````python
from random import randint
from time import sleep
````

````python
# Class Tamagotchi

class Tamagotchi:
    def __init__(self, nombre):
        self.nombre = nombre
        self.hambre = 0
        self.aburrimiento = 0
        self.cansancio = 0
        self.suciedad = 0
        self.comida = 2
        self.dormido = False
        self.vivo = True
        
    def comer(self):
        if self.comida > 0:
            self.comida -= 1
            self.hambre -= randint(1, 4)
            print(f"{self.nombre} ha comido!")
        else:
            print(f"No hay comida! {self.nombre} no puede comer...")
        
        self.hambre = 0 if self.hambre < 0 else self.hambre            
            
    def jugar(self):
        num_aleatorio = randint(0, 2)
        print(f"{self.nombre} quiere jugar... Elige un número entre 0 y 2:")
        num_usuario = input()
        if num_usuario == num_aleatorio:
            print("Correcto!")
            self.aburrimiento -= 3
        else:
            print("Incorrecto!")
            self.aburrimiento -= 1
            
        self.aburrimiento = 0 if self.aburrimiento < 0 else self.aburrimiento
            
    def dormir(self):
        self.dormido = True
        self.cansancio -= 3
        self.aburrimiento -= 2
        print(f"{self.nombre} esta durmiendo...")
        
        self.aburrimiento = 0 if self.aburrimiento < 0 else self.aburrimiento
        self.cansancio = 0 if self.cansancio < 0 else self.cansancio
        
    def despertar(self):
        num_aleatorio = randint(0, 2)
        if num_aleatorio == 0:
            print(f"{self.nombre} se ha despertado!")
            self.dormido = False
            self.aburrimiento = 0
        else:
            print(f"{self.nombre} sigue dormido.")
            self.dormir()
            
    def ducha(self):
        self.suciedad = 0
        print(f"{self.nombre} ha tomado un baño.")
        
    def buscar_comida(self):
        comida_encontrada = randint(0, 4)
        self.comida += comida_encontrada
        self.suciedad += 2
        print(f"{self.nombre} ha encontrado {comida_encontrada} comidas!")
        
    def mostrar_valores(self):
        print(f"Nombre: {self.nombre}")
        print(f"Hambre: {self.hambre}")
        print(f"Aburrimiento: {self.aburrimiento}")
        print(f"Cansancio: {self.cansancio}")
        print(f"Comida: {self.comida}")
        print(f"Suciedad: {self.suciedad}")
        print(f"Dormido: {self.dormido}")
        
    def dificultad(self, nivel):
        self.hambre += randint(0, nivel)
        self.suciedad += randint(0, nivel)
        if self.dormido == False:
            self.aburrimiento += randint(0, nivel)
            self.cansancio + randint(0, nivel)
            
    def morir(self):
        if self.hambre >= 10:
            print(f"{self.nombre} ha muerto de hambre.")
            self.vivo = False
        elif self.suciedad >= 10:
            print(f"{self.nombre} murio de una infección.")
        elif self.aburrimiento >= 10:
            print(f"{self.nombre} está muy aburrido... Se ha quedado dormido.")
            self.aburrimiento = 10
            self.dormido = True
        elif self.cansancio >= 10:
            print(f"{self.nombre} está muy cansado y se acaba de dormir.")
            self.cansancio = 10
            self.dormido = True
````

````python
# Funciones

def mostrar_menu(tamagotchi):
    if tamagotchi.dormido == False:
        print("""\n
              Presiona 1: Comer\n
              Presiona 2: Jugar\n
              Presiona 3: Dormir\n
              Presiona 4: Ducha\n
              Presiona 5: Buscar Comida
              """)
        opcion = input()
        return opcion
    else:
        print(f"{tamagotchi.nombre} está dormido... Presiona 6")
        opcion = input()
        opcion = "6" if opcion != "6" else opcion
        return opcion
    
def llamar_accion(tamagotchi, opcion):
    if opcion == "1":
        tamagotchi.comer()
    elif opcion == "2":
        tamagotchi.jugar()
    elif opcion == "3":
        tamagotchi.dormir()
    elif opcion == "4":
        tamagotchi.ducha()
    elif opcion == "5":
        tamagotchi.buscar_comida()
    elif opcion == "6":
        tamagotchi.despertar()
    else:
        print("OPCION NO VALIDA")
````

````python
# Codigo principal

print("Tamagotchi App en Python")
print("En este ejercicio vamos a intentar replicar el popular juego de tamagotchi de los 90's con python, usando clases y métodos.")

nivel_dificultad = int(input("Elije un nivel de dificultad (1 - 5):"))

nombre_tamagotchi = input("Dale un nombre al Tamagotchi: ")

mi_tamagotchi = Tamagotchi(nombre_tamagotchi)

ronda = 1

while mi_tamagotchi.vivo == True:
    print("-"*100)
    print(f"Esta es la ronda numero {ronda}")
    mi_tamagotchi.mostrar_valores()
    opcion = mostrar_menu(mi_tamagotchi)
    llamar_accion(mi_tamagotchi, opcion)
    
    mi_tamagotchi.mostrar_valores()
    
    sleep(3)
    
    mi_tamagotchi.dificultad(nivel_dificultad)
    mi_tamagotchi.morir()
    
    ronda += 1
````
